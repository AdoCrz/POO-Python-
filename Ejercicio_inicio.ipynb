#### 2. Con la ayuda de la videoconferencia de la clase y la presentación Conceptos Básicos en Machine Learning, que se encuentran en el Aula Virtual explique la diferencia entre Machine Learning y Minería de Datos.

#### Respuesta:
#### De acuerdo con lo explicado durante la videoconferencia y el material de apoyo, la principal diferencia entre Machine Learning y la Minería de Datos radica en su enfoque y en su origen, aunque ambos comparten conceptos similares.

#### La Minería de Datos se centra principalmente en aplicar técnicas estadísticas y analíticas para identificar patrones, tendencias y relaciones en conjuntos de datos completos. Su objetivo es descubrir conocimiento útil a partir de los datos existentes.

#### Por otro lado, el Machine Learning tiene un enfoque más computacional, involucrando el desarrollo y uso de algoritmos que no solo buscan identificar patrones, sino que también son capaces de aprender de los datos y replicar el comportamiento aprendido. Esto implica la capacidad de generalizar el conocimiento adquirido y hacer predicciones en nuevos datos.

#### 9. Suponga que tiene la variable "numero" la cual recibe un valor numérico. Indique con valores lógicos si el número es divisible entre 3 y entre 7. Realice una prueba para cada una de las posibilidades.

```python
# Resultado positivo alternativa 1

numero = 21
print(f'El numero {numero} puede dividirse entre 3 y 7 es: {numero % 3 == 0 and numero % 7 == 0}')
```
